# Riverbero

Ora che abbiamo capito come far funzionare l’eco, possiamo spingerci a realizzare effetti più
avanzati. Un classico è il cosiddetto **riverbero**, che simula la sensazione di diffusione del suono

cattedrale).

# Riverbero in una chiesa

Per risultati migliori, possiamo affidarci a "impulse response" preregistrati. Si tratta di registrazioni di un suono molto breve in ambienti particolari. Provate ad ascoltare questo esempio registrato in una chiesa di Nashville, Tennessee.

Il vettore ```ir``` che abbiamo ottenuto ha lo stesso ruolo del vettore ```rev``` che abbiamo definito in precedenza, nel senso che codifica un particolare effetto di riverbero su un suono ''puntiforme''. Per applicare l'effetto ad un suono qualsiasi, contenuto in un vettore ```s```, dovremo calcolare la convoluzione tra ```ir``` e ```s```. Proviamo! 

````python
# Carichiamo un suono e ascoltiamolo
s = audioread('singing.wav')
sound(s)
````

````python
# Ora applichiamo l'effetto di eco visto in precedenza.
ir = ir / sum(abs(ir))
riv = np.convolve(ir, s)
sound(riv)

# Attenzione: per un buon risultato, moltiplicate il vettore ir per un coefficiente di attenuazione, 
# ad esempio 0.1, prima di calcolarne la convoluzione con s 
````
